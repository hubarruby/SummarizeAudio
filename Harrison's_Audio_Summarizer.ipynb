```python
summarizer = pipeline("summarization", model="philschmid/bart-large-cnn-samsum")

# taken from stack exchange, splitting the text into chunks that are digestible sizes for the summarizer model
def splitter(n, s):
    pieces = s.split()
    return [" ".join(pieces[i:i+n]) for i in range(0, len(pieces), n)]

# back to code that I wrote:
split_text = splitter(700, result)
first_pass = ''
i = 0 
# summarizing chunks of the audio and concatenating them together into the first_pass variable
if input("show what's being summarized as it get's processed? (y/n): ").lower() == 'y':
  for subtext in split_text:
    print(f'subtext{i}', subtext, '\n\n'); i += 1
    first_pass += summarizer(subtext)[0].get('summary_text') + '\n'   
else:
  for subtext in split_text:
    first_pass += summarizer(subtext)[0].get('summary_text') + '\n'       
```

```python
# Your first pass summary result!
print(first_pass)
```

```python
# summarizing the summary, in case it's super long
# frankly this is not really necessary for the TEST_AUDIO file I included in the repo

split_text2 = splitter(700, first_pass)
second_pass = ''
i = 0 

if input("show what's being summarized as it get's processed? (y/n): ").lower() == 'y':
  for subtext in split_text2:
    print(f'subtext{i}', subtext, '\n\n'); i += 1
    second_pass += summarizer(subtext)[0].get('summary_text') + '\n'   
else:
  for subtext in split_text2:
    second_pass += summarizer(subtext)[0].get('summary_text') + '\n'   

print('Second Pass:\n' + second_pass)
```
